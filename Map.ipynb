```python
#Import all the modules that will be used in the program
import folium
# ==> can be installed using: pip install plotly
import pandas as pd
# ==> can be installed using pip install pandas
```

```python
#initialy I openedd the downladable dataset provided by rt.live to view the data was structured
rt_dataset = pd.read_csv("rt.csv")
rt_dataset.head()
```

```python
def get_values_for_date(given_date):
    """
    This funciton will search through the rt.live dataset and find all the rt values per state 
    for a given date
    [INPUT] ==> a date as a string in YYYY-MM-DD format: eg: 2020-11-05 
    [OUTPUT] ==> hash-map (dictionary) with the state as keys and the rt mean as the value
    """
    list_of_states_and_vals = {}
    for current_index in range(len(rt_dataset)):
        current_date = rt_dataset["date"][current_index]
        if current_date == given_date:
            current_state = rt_dataset["region"][current_index]
            current_mean_rt_value = rt_dataset["mean"][current_index]
            list_of_states_and_vals[current_state] = current_mean_rt_value
    
    return list_of_states_and_vals
```

```python
#get the infection rates for November 29th 2020
infection_rates = get_values_for_date("2020-11-29")
```

```python
def get_color(state,dict_containing_values):
    """
    This fuction will return a color for the current states as specifie don teh rt.live page
    red ==> rt value greater than 1
    green ==> rt value less than or equal to 1
    [INPUT] 1) the state you want 2) a hash-map containg all the infection rates
    [OUTPUT] a color code in RGB
    """
    if state in dict_containing_values:
        if float(dict_containing_values[state]) <= 1:
            #infection rate is less than 1
            return "#35b32e"
        else:
            #infectino rate is greater than 1
            return "#eb5358"
    else:
        #in case the current state does not match the stte code on teh geojson
        return "#808080"
    
```

```python
#create the color codes for each state
color_codes = {}
for current_state in infection_rates:
    color_codes[current_state] = get_color(current_state,infection_rates)
    
print (color_codes)
```

```python
#create the map of the USA that will be used in the analysis
covid_map = folium.Map(location = [39,-100], zoom_start=4)

# geoJson file collected from: https://eric.clst.org/tech/usgeojson/
path_to_geojson = "US_States.json"
    
folium.GeoJson(
    path_to_geojson,
    style_function=lambda feature: {
        "fillColor":get_color(state_name_to_code[feature["properties"]["NAME"]],infection_rates),
        "color": "black",
        "weight": 2,
        
    },
    highlight_function=lambda feature: {'weight':5, 'color':'black'},
).add_to(covid_map)

#add a caption title to the map
folium.LayerControl().add_to(covid_map)
covid_map
```
